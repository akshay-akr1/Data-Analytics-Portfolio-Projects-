### Task 2
The cost column in the dataframe “Project” has some missing values. Your task is to compute these missing values. Replace the missing values by running average. You should use the “For” loop for this task.

```python
# Convert the "Cost" column to numeric, forcing errors to NaN
project_dataframe_loaded["Cost"] = pd.to_numeric(project_dataframe_loaded["Cost"], errors='coerce')

# Initialize variables for running sum and count
running_sum = 0
running_count = 0

# Iterate over the DataFrame to compute the running average and replace NaNs
for i in range(len(project_dataframe_loaded)):
    # Checking if the value in the "Cost" column for the ith row of project_dataframe_loaded is not NaN (i.e., it is a valid number)
    if pd.notna(project_dataframe_loaded.loc[i, "Cost"]):
        # Update running sum and count
        running_sum += project_dataframe_loaded.loc[i, "Cost"]
        running_count += 1
    else:
        # Calculate running average and replace NaN
        if running_count > 0:
            running_average = running_sum / running_count
            project_dataframe_loaded.loc[i, "Cost"] = running_average


# Set pandas display option to show numbers in full decimal format
pd.set_option('display.float_format', '{:.2f}'.format)

# Displaying the updated Dataframe
project_dataframe_loaded
```

### Task 3
Split the name column in the Employee dataframe into two new columns “First Name”, and “Last Name” and remove the older “name” column.

```python
# Split the 'Name' column into 'First Name' and 'Last Name'
employee_dataframe_loaded[['First Name', 'Last Name']] = employee_dataframe_loaded['Name'].str.split(' ', n=1, expand=True)

# Drop the old 'Name' column
employee_dataframe_loaded = employee_dataframe_loaded.drop(columns=['Name'])

# Displaying updated Dataframe
employee_dataframe_loaded
```

### Task 4
Join all three dataframes in one single dataframe. Name it “Final”

```python
# Merging project_dataframe_loaded with employee_dataframe_loaded on the "ID" column
merged_df1 = pd.merge(left=project_dataframe_loaded, right=employee_dataframe_loaded, on = "ID")

# Merging the resulting DataFrame (merged_df1) with seniority_level_dataframe_loaded on the "ID" column
Final = pd.merge(left=merged_df1,right=seniority_level_dataframe_loaded, on = "ID")

# Displaying the final merged DataFrame
Final
```

### Task 5
Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects.

```python
# Convert 'Cost' column to numeric type
Final['Cost'] = pd.to_numeric(Final['Cost'], errors='coerce')

# Initialize the 'Bonus' column to 0
Final['Bonus'] = 0

# Calculate the bonus (5% of Cost) for finished projects
Final.loc[(Final['Status'] == 'Finished'), 'Bonus'] = Final['Cost'] * 0.05

# Displaying the updated final merged DataFrame
Final
```

### Task 6
Demote the designation level by 1, whose projects have status “fail”. Delete the
employees record whose designation level is above 4.

```python
# Demote the designation level by 1 for employees with failed projects
Final.loc[Final['Status'] == 'Failed', 'Designation Level'] = Final['Designation Level'] - 1

# Delete records where the 'Designation Level' is above 4
Final = Final[Final['Designation Level'] <= 4]

# Displaying the updated final merged DataFrame
Final 
```

### Task 7
Add “Mr.” and “Mrs.” to the first name column and drop the gender column.

```python
# Add "Mr." or "Mrs." to the First Name based on Gender using np.where
Final['First Name'] = np.where(Final['Gender'] == 'M', 'Mr. ' + Final['First Name'], 'Mrs. ' + Final['First Name'])

# Drop the Gender column
Final = Final.drop(columns=['Gender'])

# Displaying the updated final merged DataFrame
Final 
```

### Task 8
Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.

```python
# Promote designation level by 1 for employees whose age is more than 29 years
Final.loc[Final['Age'] > 29, 'Designation Level'] += 1

# Displaying the updated final merged DataFrame
Final 
```

### Task 9
Add the cost of all projects for each Employee and save it in new dataframe
“TotalProjCost” with three columns ID, First Name, and Total cost

```python
# Group by 'ID' and sum the 'Cost' for each employee
total_cost = Final.groupby('ID')['Cost'].sum().reset_index()

# Rename the 'Cost' column to 'Total Cost'
total_cost.rename(columns={'Cost': 'Total Cost'}, inplace=True)

# Merge with employee details to get 'First Name'
total_cost = total_cost.merge(Final[['ID', 'First Name']].drop_duplicates(), on='ID', how='left')

# Create the TotalProjCost DataFrame
TotalProjCost = total_cost[['ID', 'First Name', 'Total Cost']]

# Displaying TotalProjCost DataFrame
TotalProjCost
```
